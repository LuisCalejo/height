## Self-reported Height (CDC Survey)

```python
file_dir_sap = '../data/LLCP2023.XPT' 
# data can be downloaded at https://www.cdc.gov/brfss/annual_data/annual_2023.html
```

```python
df2 = pd.read_sas(file_dir_sap, format="xport")
df2["_SEX"].value_counts().sort_index().reset_index()
```

```python
# Filter data
# only adults
# only male
# answered in imperial measuring system

print(df2["_SEX"].value_counts().sort_index())
print("REMOVE OTHER SEX:")
df2["LADULT1"]
df2=df2[df2["_SEX"]==gender]
print(df2["_SEX"].value_counts().sort_index())

```

```python
df2["HEIGHT3"].value_counts().sort_index()
```

```python
# remove missing data
print("Removed missing data")
df2 = df2[df2["HEIGHT3"]!=7777] 
df2 = df2[df2["HEIGHT3"]!=9999] 


len(df2)
```

```python
 #only heights between 4 and 7 feet (and keep 9 for the metric system)
df2 = df2.loc[df2["HEIGHT3"].astype(str).str[0].isin(['4', '5', '6', '7', '9'])]

# metric system: only height up to 2.20 meters
df2 = df2[df2["HEIGHT3"]<=9220] 


len(df2)
```

```python
 #remove invalid imperial system entries with 12 inches
df2 = df2.loc[~((df2["HEIGHT3"].astype(str).str[1:3] == '12') & (df2["HEIGHT3"].astype(str).str[0] != '9'))]
len(df2)
```

```python
df2["height_cm"] = [
    int(str(x)[1:4]) if str(x)[0] == '9'  # If first digit is '9', it's already in cm
    else int(str(x)[0]) * 30.48 + int(str(x)[1:3]) * 2.54  # Convert imperial to cm
    for x in df2["HEIGHT3"]
]
```

```python
#unweighted average height
df2["height_cm"].mean() 
```

```python
df2["height_inches_rounded"] = [int(x/2.54)for x in df2["height_cm"]]
```

```python
# # Frequency by height without weighting
# df2["height_inches_rounded"].value_counts().to_csv("height_selfreported_unweighted.csv", index=False)
# df2["height_inches_rounded"].value_counts().sort_index()
```

```python
# Frequency by height without weighting
df2.groupby("height_inches_rounded").size().reset_index(name="count").to_csv("height_selfreported_unweighted.csv", index=False)
df2.groupby("height_inches_rounded").size().reset_index(name="count")
```

```python
# weighted average height
(df2["height_cm"] * df2["_LLCPWT"]).sum() / df2["_LLCPWT"].sum()
```

```python
# Multiply frequency by weights (_LLCPWT) to get weighted distribution.
df2.groupby("height_inches_rounded")["_LLCPWT"].sum().reset_index().to_csv("height_selfreported_weighted.csv", index=False)
df2.groupby("height_inches_rounded")["_LLCPWT"].sum().reset_index()
```

```python

# Demographics
file_dir_dem = '../data/DEMO_L.xpt'
df_dem = pd.read_sas(file_dir_dem, format="xport")

# Measured heights:
file_dir_height = '/Users/luis/Downloads/BMX_L.xpt'
df_height = pd.read_sas(file_dir_measured, format="xport")
```

```python
df_dem["RIAGENDR"].value_counts().sort_index().reset_index()
```

```python
# filter based on age and sex
df_dem = df_dem[df_dem["RIDAGEYR"]>=19] # at least 19 years old
df_dem = df_dem[df_dem["RIAGENDR"]==gender] # male
```

```python
# Perform inner join on SEQN
df_merged = pd.merge(df_dem, df_height, on="SEQN", how="inner")
```

```python
# drop NAs from df_merged for column BMXHT
df_merged = df_merged.dropna(subset=["BMXHT"]).reset_index(drop=True)
```

```python
# df_merged["BMXHT"]
```

```python
# Convert cm to total inches
df_merged["total_inches"] = df_merged["BMXHT"] * 0.393701
df_merged["total_inches_rounded"] = df_merged["total_inches"].round()

# Calculate feet and remaining inches
df_merged["feet"] = df_merged["total_inches"] // 12
df_merged["inches"] = df_merged["total_inches"] % 12

# Format the string
df_merged["feet inches"] = df_merged["total_inches"].round().astype(int).astype(str) + " - " + \
                          df_merged["feet"].astype(int).astype(str) + " ft " + \
                          df_merged["inches"].round().astype(int).astype(str)

```

```python
# Average height (unweighted)
df_merged["total_inches"].mean()*2.54
```

```python
import numpy as np
import matplotlib.pyplot as plt

unique_values = np.unique(df_merged["total_inches_rounded"])

plt.figure(figsize=(12, 6))
df_merged["total_inches_rounded"].hist(
    bins=unique_values, 
    weights=df_merged["WTMEC2YR"],  # Apply weights
    align='left', 
    rwidth=0.8
)

plt.title('Weighted Distribution of Heights')
plt.xlabel('Height (inches)')
plt.ylabel('Weighted Frequency')
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# Export weighted histogram
height_weighted_counts = df_merged.groupby("total_inches_rounded")["WTMEC2YR"].sum().reset_index()

# Rename columns
height_weighted_counts.columns = ["Height (inches)", "Total Weighted Count"]

# Export to CSV
height_weighted_counts.to_csv("height_measured_weighted.csv", index=False)

# Display first rows
print(height_weighted_counts.head())
```

```python
# Export histogram
height_distribution = df_merged["total_inches_rounded"].value_counts().sort_index()

# Convert to DataFrame
height_distribution_df = height_distribution.reset_index()
height_distribution_df.columns = ["Height (inches)", "Frequency"]

# Export to CSV
height_distribution_df.to_csv("height_measured_unweighted.csv", index=False)

# Display first few rows
print(height_distribution_df.head())
```
